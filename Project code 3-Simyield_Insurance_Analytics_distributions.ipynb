```python
import pandas as pd
import numpy as np

# read the input file generated for running insurance anlytics and filter the yield values in df2 dataframe
df1=pd.read_csv("C:/Madhur/APY_NDVI_1/Rice_AP_250mnomask6.csv")
filter_col = [col for col in df1 if col.startswith('stat ')]
df2=df1[filter_col]
df2
```

```python
# read the threshold yield file created and read the threshold values in df data frame

df=pd.read_csv("C:/Madhur/APY_NDVI_1/Rice_AP_TY_APY_mean_3.csv")
filter_col = [col for col in df if col.startswith('TY')]
df=df[filter_col]
#df=df*1000
df
```

```python
# calculating the burn rates in percentage by subtracting the yields from threshold yields and then dividing and discarding the values with no loss

df3=pd.DataFrame((np.array(df)-np.array(df2)))
df3=pd.DataFrame((np.array(df3)/np.array(df)))
#df3=df3/df
df3[df3<0]=0
df3['Concat']=df1['Concat']
df3
```

```python
# calculated the uncapped burn rates and average of uncapped burn rates ( this fragment represent the earlier scenario of capped and uncapped burn rates which was dropped later and replaces with area weighterd pure premium rates which is calculated in next steps)
# can comment out this block if not required for analysis
uncap=df3
uncap_mean=uncap.mean(axis=1)
uncap["Mean Uncapped"]=uncap_mean
uncap
purepremiumrate=pd.DataFrame(uncap_mean)
#uncap_90=uncap.quantile(0.9,axis=1)
#uncap["Uncapped Burn Cost"]=uncap_mean
#cap=np.minimum(uncap, uncap_90[:,None])
#cap['90perc']=uncap_90
```

```python
#calculation of Sum Insured ( Replace MSP per crop manually)

MSP=1470
Si=df.multiply(MSP/100)
Si
```

```python
#calculation of pure premium rate district wise 

dist_list=list(df1.Concat.unique())
del df3["Mean Uncapped"]
pp=[]

for i in dist_list:
    df5=df1[df1['Concat']==i]
    filter_col = [col for col in df5 if col.startswith('Area')]
    area=df5[filter_col]
    df4=df3[df3['Concat']==i]
    del df4["Concat"]
    e=pd.DataFrame(np.array(area)*np.array(df4))
   
    pp1=e.sum().sum()/area.sum().sum()
    pp.append(pp1)
pp
```

```python
# repeating the same pure premium for all bins over the district
pp=pd.DataFrame(pp)
pp=pp.loc[pp.index.repeat(20)]
```

```python
#calculating state wide purepremium rate

del df3["Concat"]
filter_col = [col for col in df1 if col.startswith('Area')]
area=df1[filter_col]
e=pd.DataFrame(np.array(area)*np.array(df3))
pr_areaweighted=e.sum().sum()/area.sum().sum()
pr_areaweighted
```

```python
# calculating Total claims (Sum Insure X Burn Rates X Area)

tl=(np.array(Si)*np.array(df3))
tl=pd.DataFrame(tl)
tl=(np.array(tl)*np.array(area))/10000000
tl=pd.DataFrame(tl)
tl
```

```python
#calculating total premiums (Sum Insured X Area X Pure Premium Rate)

o=pd.DataFrame(np.array(Si)*np.array(area))
l=pd.DataFrame(np.array(o)*np.array(pp))/10000000
l
```

```python
# stoing and writing the final output file *make sure to change the m.column names as per analysis for correct formatting
df1.reset_index(drop=True, inplace=True)
df.reset_index(drop=True, inplace=True)
df3.reset_index(drop=True, inplace=True)
pp.reset_index(drop=True, inplace=True)
tl.reset_index(drop=True, inplace=True)
l.reset_index(drop=True, inplace=True)
m=pd.concat([df1,df,df3,pp,tl,l],axis=1,ignore_index=True)
```
